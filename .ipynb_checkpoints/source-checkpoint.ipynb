```python
import cv2
import sklearn
import os
from sklearn.cluster import KMeans
import time, sys
import numpy as np
import pickle
from matplotlib import pyplot as plt
from IPython.display import clear_output
```

```python
def printProgressBar (iteration, total, prefix = '', suffix = '', decimals = 1, length = 100, fill = '█', printEnd = "\r"):
    """
    Call in a loop to create terminal progress bar
    @params:
        iteration   - Required  : current iteration (Int)
        total       - Required  : total iterations (Int)
        prefix      - Optional  : prefix string (Str)
        suffix      - Optional  : suffix string (Str)
        decimals    - Optional  : positive number of decimals in percent complete (Int)
        length      - Optional  : character length of bar (Int)
        fill        - Optional  : bar fill character (Str)
        printEnd    - Optional  : end character (e.g. "\r", "\r\n") (Str)
    """
    percent = ("{0:." + str(decimals) + "f}").format(100 * (iteration / float(total)))
    filledLength = int(length * iteration // total)
    bar = fill * filledLength + '-' * (length - filledLength)
    print('\r%s |%s| %s%% %s' % (prefix, bar, percent, suffix), end = printEnd)
    # Print New Line on Complete
    if iteration == total: 
        print()
```

```python
data_path = './oxford/images'
img_train = []
num_data = len(os.listdir(data_path))
filename = []
for idx, val in enumerate(os.listdir(data_path)):
    img = cv2.imread(os.path.join(data_path, val), cv2.IMREAD_GRAYSCALE)
    img_train.append(img)
    filename.append(val)
    printProgressBar(idx + 1, num_data, prefix = 'Load images:', suffix = 'Complete', length = 50)
```

```python
sift = cv2.xfeatures2d.SIFT_create()

def features(img, extractor):
    keypoints, descriptors = sift.detectAndCompute(img, None)
    return keypoints, descriptors
```

```python
def extract_feature(img_list):
    global filename
    descriptor_list = []
    num_img = len(img_list)
    for idx, img in enumerate(img_list):
        try:
            kp, des = features(img, sift)
            descriptor_list.append(des)
            printProgressBar(idx + 1, num_img, prefix = 'Extract features:', suffix = 'Complete', length = 50)
        except:
            print(filename[idx])
    return descriptor_list
descriptor_list = extract_feature(img_train)
```

```python
def build_histogram(descriptor_list, cluster_alg):
    histogram = np.zeros(len(cluster_alg.cluster_centers_))
    cluster_result =  cluster_alg.predict(descriptor_list)
    for i in cluster_result:
        histogram[i] += 1.0
    return histogram
```

```python
wcss = []
for i in range(1,45):
    k = 250*i
    fname = './mini_batch_model/kmeans_'+str(k)+'.sav'
    file = open(fname, 'rb')
    kmeans = pickle.load(file)
    file.close()
    wcss.append(kmeans.inertia_)
fig = plt.figure()
plt.plot([250*i for i in range(1,45)], wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
fig.savefig('elbow.svg', bbox_inches='tight')
print(wcss)
```

```python
preprocessed_image = []
fname = './mini_batch_model/kmeans_5000.sav'
file = open(fname, 'rb')
kmeans = pickle.load(file)
file.close()
for idx, descriptor in enumerate(descriptor_list):
    clear_output(wait=True)
    if (descriptor is not None):
        histogram = build_histogram(descriptor, kmeans)
        preprocessed_image.append(histogram)
        pickle.dump(preprocessed_image, open('preprocessed_image.sav', 'wb'))
    printProgressBar(idx + 1, len(descriptor_list), prefix = 'Histogram:', suffix = 'Complete', length = 50)
```

```python
# load running time
time_file = './checkpoint/running_time.sav'
file = open(time_file, 'rb')
running_time = pickle.load(file)
fig = plt.figure()
plt.plot([250*i for i in range(1,46)], running_time)
plt.title('Training time')
plt.xlabel('Number of clusters')
plt.ylabel('Seconds')
plt.show()
fig.savefig('running_time.svg', bbox_inches='tight')
```
